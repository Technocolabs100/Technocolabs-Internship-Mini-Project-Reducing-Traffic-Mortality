```python
#2

import pandas as pd

# Read 
accidents = pd.read_csv('road-accidents.csv', comment='#', sep ='|')

# Save the number of rows columns
rows_columns = accidents.shape
print('Number of rows are: ')
print(rows_columns[0])
print('------------------------------------------------------------')
print('Number of columns are: ')
print(rows_columns[1])
print('------------------------------------------------------------')

#overview
car_accident_info =  accidents.info()
print(car_accident_info)

# DataFrame
accidents.head()
accidents.tail()
```

```python
#3

import seaborn as sns
%matplotlib inline

# Compute the summary statistics of all columns
stat=accidents.describe()
print(stat)

# pairwise scatter plot
sns.pairplot(accidents)
```

```python
#4

# Correlation coefficent for all column pairs
correlation_coefficient = accidents.corr()
correlation_coefficient
```

```python
#5

from sklearn import linear_model

# Creating features and target DataFrames

f = accidents.drop(['state','drvr_fatl_col_bmiles'], axis =1)
t =  accidents['drvr_fatl_col_bmiles']

# Create a linear regression object
regression = linear_model.LinearRegression()

# Fit a multivariate linear regression model
regression.fit(f, t)

# Retrieve the regression coefficients
result = regression.coef_
result
```

```python
#6

# Standardize and center the feature columns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
measure = StandardScaler()
features_scaled = measure.fit_transform(f)

# Import the PCA class function from sklearn
analysis = PCA()

# Fit the standardized data to the pca
common=analysis.fit(features_scaled)
common

# Plot the proportion of variance explained on the y-axis of the bar plot
import matplotlib.pyplot as plt
plt.xlabel('X_Principal components ')
plt.ylabel('Y_Proportion of variance')
plt.xticks([1, 2, 3])
plot=plt.bar(range(1, analysis.n_components_ + 1),  analysis.explained_variance_ratio_, color = ['blue', 'pink'])
plot

# Compute the cumulative proportion of variance explained by the first two principal components
components = analysis.explained_variance_ratio_.cumsum()[1]
j=round(components, 5)
print("The cumulative variance of the first two principal components is:")
print(j)
```

```python
#7

# Transform the scaled features using two principal components
analysis = PCA(n_components=2)
principal_components = analysis.fit_transform(features_scaled)

# Extract the first and second component to use for the scatter plot
p_comp1 = principal_components[:,0]
p_comp2 = principal_components[:,1]

# plot title
plt.title(' Scatter plot of first two principal components ',color = 'blue')

# Plot the first two principal components in a scatter plot
plt.scatter(p_comp1,p_comp2,color = 'red')
plt.show()
```

```python
#8

# Import KMeans from sklearn
from sklearn.cluster import KMeans

# A loop will be used to plot the explanatory power for up to 10 KMeans clusters
ks = range(1, 10)
inertias = []
for k in ks:
    # Initialize the KMeans object using the current number of clusters (k)
    km = KMeans(n_clusters= k, random_state=8)
    # Fit the scaled features to the KMeans object
    km.fit(features_scaled)
    # Append the inertia for `km` to the list of inertias
    inertias.append(km.inertia_)
    
# Plot the results in a line plot
plt.plot(ks, inertias, marker='o')
plt.show()
```

```python
#9

# Create a KMeans object with 3 clusters, use random_state=8 
km = KMeans(n_clusters = 3, random_state=8)

# Fit the data to the `km` object
km.fit(features_scaled)

# Create a scatter plot of the first two principal components
# and color it according to the KMeans cluster assignment 
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=km.labels_)
plt.show()
```

```python
#10

# Create a new column with the labels from the KMeans clustering
accidents['cluster'] = km.labels_

# Reshape the DataFrame to the long format
melt_car = pd.melt(accidents, id_vars = ['cluster'], var_name = 'measurement', value_name = 'percent', value_vars = f.columns)

# Create a violin plot splitting and coloring the results according to the km-clusters
sns.violinplot(melt_car['percent'],melt_car['measurement'], hue = melt_car['cluster'])
```

```python
#11

# Read in the new dataset
miles_driven = pd.read_csv('miles-driven.csv', sep='|')


car_acc_miles = accidents.merge(miles_driven, on ='state')

a=car_acc_miles['drvr_fatl_col_bmiles']/1000
b=car_acc_miles['million_miles_annually']
c=display(car_acc_miles.head())


car_acc_miles['num_drvr_fatl_col']=a*b


# Create a barplot of the total number of accidents per cluster
sns.barplot(x="cluster", y="num_drvr_fatl_col", data=car_acc_miles, estimator=sum, ci=None)

# Calculating the number of states in each cluster and their  mean and sum.
count_mean_sum = car_acc_miles.groupby('cluster')[['num_drvr_fatl_col']].agg(['count','mean','sum'])
count_mean_sum
```
